# World Happiness Report 2021 Dataset Analysis

### About World Happiness Report

The World Happiness Report is an annual publication that measures and assesses the well-being and happiness of people in countries around the globe. This comprehensive study, first published in 2012 by the Sustainable Development Solutions Network (SDSN) in conjunction with the United Nations, provides valuable insights into factors influencing happiness and overall life satisfaction.

The report evaluates several key elements, such as income levels, social support systems, life expectancy, freedom to make life choices, generosity, and perceptions of corruption. It aims to promote a better understanding of happiness as a crucial component of societal progress and development.

#### Renaming The Columns

```python
df.rename(columns={"Country name":"country_name","Regional indicator":"regional_indicator"
            ,"Ladder score":"happiness_score","Logged GDP per capita":"logged_GDP_per_capita"
            ,"Social support":"social_support","Healthy life expectancy":"healthy_life_expectancy"
            ,"Freedom to make life choices":"freedom_to_make_life_choices","Generosity":"generosity"
            ,"Perceptions of corruption":"perceptions_of_corruption"},inplace=True)
```

```python
df.head()
```

### Extracting Useful Columns

```python
df_columns = ["country_name", "regional_indicator", "happiness_score", "logged_GDP_per_capita", "social_support"
             ,"healthy_life_expectancy","freedom_to_make_life_choices", "generosity", "perceptions_of_corruption"]
```

```python
df = df[df_columns].copy()
```

```python
df.head()
```

```python
# Checking null values

df.isnull().sum()
```

There are no null values in this dataset.

### Plot b/w Happines Score and GDP per capita

```python
plt.figure(figsize=(15,8))
plt.title("Plot Happiness score vs GDP")
sns.scatterplot(x = df.happiness_score, y = df.logged_GDP_per_capita, hue = df.regional_indicator
               ,s = 200)
plt.legend(loc = "upper left", fontsize="10")
plt.xlabel("Happiness Score")
plt.ylabel("GDP Per Capita")
plt.show()
```

```python
gdp_region = df.groupby("regional_indicator")["logged_GDP_per_capita"].sum().sort_values(ascending=False)
gdp_region
```

### GDP By Region Pie Chart

```python
plt.figure(figsize=(12,8))
plt.title("GDP By Region",fontsize=15)
p = plt.pie(gdp_region,labels = df.regional_indicator.value_counts().index, autopct = '%1.1f%%')
plt.show()
```

### Region wise Number of Countries

```python

total_country = df.groupby("regional_indicator")["country_name"].count().sort_values(ascending=False)
total_country
```

```python
plt.figure(figsize = (12,8))
sns.countplot(y='regional_indicator', order = df['regional_indicator'].value_counts().index[:],data = df)
plt.title('Region wise Number of countries',fontsize=20)
plt.ylabel("Regions")
plt.xlabel("No. of Countries")
plt.show()
```

### Correlation map

```python
cor = df.corr(method = "pearson")
plt.subplots(figsize=(10,5))
sns.heatmap(cor,cmap = "Blues", square = True)
plt.show()
```

### Corruption In Different Regions

```python
corruption = df.groupby("regional_indicator")[["perceptions_of_corruption"]].mean().sort_values(by="perceptions_of_corruption",ascending=False)
corruption
```

```python
plt.figure(figsize = (12,8))
plt.title("Perception Of Corruption In Various Regions",fontsize=17)
plt.xlabel("Regions",fontsize=15)
plt.ylabel("Corruption Index",fontsize=15)
plt.xticks(rotation=30, ha="right")
plt.bar(corruption.index, corruption.perceptions_of_corruption)
plt.show()
```

## Country Wise Life Expectancy Plots

```python
top_10 = df.head(10)
bottom_10 = df.tail(10)
```

```python
plt.figure(figsize=(12,8))
plt.title("Top 10 Happiest Countries Life Expectancy",fontsize=20)
sns.barplot(x = top_10.country_name, y = top_10.healthy_life_expectancy)
plt.xlabel("Country Name",fontsize=15)
plt.ylabel("Life Expectancy",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Bottom 10 Happiest Countries Life Expectancy",fontsize=20)
sns.barplot(x = bottom_10.country_name, y = bottom_10.healthy_life_expectancy)
plt.xlabel("Country Name",fontsize=15)
plt.ylabel("Life Expectancy",fontsize=15)
plt.show()
```

### Happiness Score vs Freedom to Make Life Choices Plot

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = df.freedom_to_make_life_choices, y = df.happiness_score, hue = df.regional_indicator
               ,s = 200)
plt.legend(loc = "upper left",fontsize=12)
plt.xlabel("Freedom To Make Life Choices")
plt.ylabel("Happiness Score")
plt.show()
```

### Top 10 Corrupt Countries

```python
country = df.sort_values(by = "perceptions_of_corruption").tail(10)
plt.figure(figsize=(15,8))
plt.title("Countries with Most Perception Of Corruption",fontsize=20)
plt.xlabel("Country", fontsize=15)
plt.ylabel("Corruption Index", fontsize=15)
plt.xticks(rotation=30)
plt.bar(country.country_name, country.perceptions_of_corruption, color="maroon")
plt.show()
```

### Variation of Happiness Score with Corruption

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = df.happiness_score, y = df.perceptions_of_corruption, hue=df.regional_indicator
               ,s=200)
plt.title("Happines Score vs Corruption",fontsize="15")
plt.legend(loc="lower left", fontsize="15")
plt.xlabel("Happiness Score",fontsize=15)
plt.ylabel("Corruption",fontsize=15)
plt.show()
```
